```python
import pandas as pd

# 讀取Excel檔案
file_path = '/content/train0731.xlsx'
df = pd.read_excel(file_path)

# 將NaN替換為空字串並移除換行符，然後合併指定的文字欄位
df['Merged_Info'] = (df['ComProfile'].fillna('') + ' ' +
                     df['Business_Item_Desc'].fillna('') + ' ' +
                     df['ProductInfo'].fillna('') + ' ' +
                     df['Business_Item_eTax'].fillna('') + ' ' +
                     df['product/service'].fillna('')).replace('\n', ' ', regex=True)

# 合併Bom_label欄位
df['Merged_Bom_labels'] = df['Bom_label_1'].fillna('')

# 保留指定的列
df_final = df[['CompanyName', 'Merged_Info', 'Merged_Bom_labels']]

# 儲存修改後的檔案
output_path = '修改後的檔案.xlsx'
df_final.to_excel(output_path, index=False)

print('檔案已成功修改並儲存!')
```

```python
import pandas as pd
import json

def excel_to_json(file_path, output_path):
    # 讀取Excel檔案
    df = pd.read_excel(file_path)

    # 構建JSON結構
    data = []
    for _, row in df.iterrows():
        item = {
            "instruction": "Please answer which industry category number the following should be.",
            "input": row['Merged_Info'],
            "output": row['Merged_Bom_labels']
        }
        data.append(item)

    # 將數據寫入JSON檔案
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    print(f'JSON file has been saved to {output_path}')

# 定義檔案路徑
file_path= '/content/修改後的檔案.xlsx'  # 更改為你的檔案路徑
output_path = 'output0731.json'  # 指定你想保存的JSON檔案路徑

# 執行函數
excel_to_json(file_path, output_path)
```
